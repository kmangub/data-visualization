# Data Visualization B: Hotel Bookings

The Hotel Bookings dataset shows information regarding hotel booking and stays.

```
df = pd.read_csv('hotel_bookings.csv')
df = df[df['adr'] < 1000 ]
df.head(20)
```

Fo our data visualization:
 - We will be using the **leadtime** column, which is the number of days that elapsed between the date booked and the arrival date.
 - We will also use the **adr** column, which tells us the average daily rate.

 Plotting our x and y values:

```
g = sns.relplot(x= 'lead_time', y = 'adr', data = df)
g.set(xlabel='Lead Time', ylabel='Average Daily Rate in Dollars')

print()
```

Looking at our graph above answers the question, does booking a hotel affect the the cost? The answer is yes.

The lead time for booking will result in a cheaper average daily rate. As the lead time approaches 0, otherwise known as checking in to the hotel the day of, it will cost us more money.
